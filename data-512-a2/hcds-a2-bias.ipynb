```python
import requests
headers = {
    'User-Agent': 'https://github.com/kevSweet',
    'From': 'ksweet1@uw.edu'
}

def api_call(endpoint,revids):
    call = requests.get(endpoint + revids, headers=headers)
    response = call.json()
    return response
```

```python
ores_base_url = 'https://ores.wikimedia.org/v3/scores/enwiki?models=articlequality&revids='
rev_ids = '355319463|393276188'
```

```python
test = api_call(ores_base_url, rev_ids)
test
```

```python
rev_strings = page['rev_id'].values.astype(str)
rev_strings
```

```python
def retrieve_data(rev_strings, points=50):
    data = []
    start = 0
    stop = points
    while len(page[start:stop]) > 0:
        data.append(api_call(ores_base_url, '|'.join(rev_strings[start:stop])))
        start += points
        stop += points
    return data

predictions = retrieve_data(rev_strings)
predictions
```

```python
scored = pd.DataFrame(columns=['rev_id', 'prediction'])
not_scored = pd.DataFrame(columns=['rev_id'])

for query in predictions:
    for value, key in query['enwiki']['scores'].items():
        if key['articlequality'].get('score') is not None:
            prediction = key['articlequality']['score']['prediction']
            scored = scored.append({'rev_id': value, 'prediction': prediction}, ignore_index=True)
        else:
            not_scored = not_scored.append({'rev_id': value}, ignore_index=True)
```

```python
scored['rev_id'] = pd.to_numeric(scored['rev_id'])
scored
```

```python
not_scored
```

```python
not_scored.to_csv('not_scored_log.csv')
```

```python
merged = pd.merge(page, scored, on=['rev_id'], how='inner')
merged = pd.merge(merged, country, left_on='country', right_on='Name', how='outer')
merged
```

```python
nomatch = merged[merged.isna().any(axis=1)]
nomatch
```

```python
nomatch.to_csv('wp_wpds_countries-no_match.csv')
```
